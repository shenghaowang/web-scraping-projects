# Collect Douban movie data with BeautifulSoup

```python
import urllib.request as urlrequest
from bs4 import BeautifulSoup
```

urllib.request: Python module used for fetching URLs (Uniform Resource Locators);

Beautiful Soup: Python library for pulling data out of HTML and XML files.

```python
douban_url = 'https://movie.douban.com/subject/3319755/'
page = urlrequest.urlopen(douban_url).read() # Open a network object denoted by a URL for reading.
#print(page)
soup = BeautifulSoup(page, 'html.parser') # Read html from network object
print(soup)
```

```python
movie = soup.find('h1').span.get_text()
#movie = soup.find('h1').find_all('span')[1].get_text()
rating = soup.find(class_='ll rating_num').get_text()
director = soup.find(id='info').find_all('span')[0].get_text()
date = soup.find(id='info').find_all('span')[16].get_text()
```

```python
print('{}, {}, {}, {}'.format(movie, rating, director, date))
```

# Collect data of Douban Top 250 books

```python
from lxml import etree
import requests
import time
```

### Trial of fetching data of top 25 books

```python
url = 'https://book.douban.com/top250?start=[]'.format(0)
print(url)
data = requests.get(url).text
print(data)
s = etree.HTML(data)
#print(s)
file = s.xpath('//*[@id="content"]/div/div[1]/div/table') # Collect book information in table format
#print(file)
```

### Actual implementation

```python
# Iterate over 10 pages with 25 books on each page
for a in range(10):
    # url is updated for each page, i.e. ... start=0, 25, 50...
    url = 'https://book.douban.com/top250?start=[]'.format(a*25)
    data = requests.get(url).text
    
    # Collect book information in table format
    s = etree.HTML(data)
    file = s.xpath('//*[@id="content"]/div/div[1]/div/table')
    time.sleep(3)
    
    for div in file:
        book = div.xpath("./tr/td[2]/div[1]/a/@title")[0]
        link = div.xpath("./tr/td[2]/div[1]/a/@href")[0]
        rating = div.xpath("./tr/td[2]/div[2]/span[2]/text()")[0]
        comments = div.xpath("./tr/td[2]/div[2]/span[3]/text()")[0].strip("(").strip().strip(")").strip()
        theme = div.xpath("./tr/td[2]/p[2]/span/text()")
        
        if len(theme) > 0:
            print("{}, {}, {}, {}, {}\n".format(book, link, rating, comments, theme[0]))
        else:
            print("{}, {}, {}, {}\n".format(book, link, rating, comments))
```
